#Import Necessary Libraries

```python
import numpy as np
import gym
env= gym.make('FrozenLake-v0')
env.render()
```

# Value Iteration

# Computing the optimal value function

```python
num_iterations=1000
threshold=1e-10
gamma=0.9
value_table=np.zeros(env.observation_space.n)
V=np.zeros(num_iterations)
Error=np.zeros(num_iterations)
for i in range(num_iterations):
  updated_value_table=np.copy(value_table)
  for s in range(env.observation_space.n):
    Q_values=[sum([prob*(r+gamma*value_table[s_]) 
                       for prob,s_,r,_ in env.P[s][a]]) 
                            for a in range(env.action_space.n)]
    value_table[s]=max(Q_values)
  V[i+1]=np.sum(value_table)
  Error[i]=np.sum(np.abs(updated_value_table-value_table))
  if (Error[i]<=threshold):
    break
Total_iterations=i                                                            
```

# Checking Convergence Speed

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
plt.plot(V[0:Total_iterations])
plt.xlabel('Number of Iterations k')
plt.ylabel('sum of $v_k(s)$ for all state $s$')
plt.title("Discount Factor 0.9")
plt.savefig("Value_Iteration.jpg")
```

```python
import pylab
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 14})
fig=plt.plot(Error[0:Total_iterations])
plt.yscale('log')
plt.xlabel('Number of Iterations k')
plt.ylabel('$Error_k$')
plt.title("Discount Factor 0.9")
plt.savefig("Value_Iteration.jpg")
```

# Extracting the optimal policy

```python
policy=np.zeros(env.observation_space.n)
for s in range(env.observation_space.n):
  Q_values=[sum([prob*(r+gamma*value_table[s_]) for prob, s_, r, _ in env.P[s][a]])
                                                     for a in range(env.action_space.n)]
  policy[s]=np.argmax(np.array(Q_values))                                                  
```
